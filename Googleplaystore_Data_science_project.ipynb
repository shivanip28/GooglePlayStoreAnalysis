```
 datamodel = dataset.drop(["Category","Size","Price", "Type", "Content Rating", "Genres", "App", "Last Updated", "Current Ver", "Android Ver"], axis =1)
```

```
datamodel.head()
```

**IMPORTING LIBRARIES FOR DATA MODELING**

```
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
lr = LinearRegression()
```

**TRAINING THE DATASET USING test_train_split**

```
X = datamodel.loc[:, datamodel.columns != "Rating"]
y = datamodel['Rating']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 10)
```

```
model_lr = lr.fit(X_train, y_train)
y_predict = model_lr.predict(X_test)
y_predict
```

```
 predicted= pd.DataFrame(data={"Predicted": y_predict, "Actual": y_test})
predicted.head(10)
```

```
  predicted[:20].plot(kind = "bar", figsize = (10,8))
```

```
lr_score=model_lr.score(X_test, y_test)
lr_score
```

```
mse_lr = np.mean((y_predict - y_test)**2)
mse_lr
```

```

 
svr = SVR()
 
model_svr = svr.fit(X_train, y_train)
 
y_predict_svr = model_svr.predict(X_test)
y_predict_svr
```

```
predicted= pd.DataFrame(data = {"Predicted": y_predict_svr, "Actual": y_test})
predicted
```

```
predicted[:20].plot(kind = "bar", figsize = (10,8))
```

```
svr_score=model_svr.score(X_test, y_test)
svr_score
```

```
 mse_svr = np.mean((y_predict_svr - y_test)**2)
mse_svr
```

```

rfr = RandomForestRegressor()
model_rfr = rfr.fit(X_train, y_train)
 
y_predict_rfr = model_rfr.predict(X_test)
y_predict_rfr
```

```
predicted= pd.DataFrame(data={"Predicted": y_predict_rfr, "Actual": y_test})
predicted
```

```
predicted[:20].plot(kind = "bar", figsize = (10,8))
```

```
rfr_score= model_rfr.score(X_test, y_test)
rfr_score
```

```
 mse_rfr= np.mean((y_predict_rfr - y_test)**2)
mse_rfr
```
